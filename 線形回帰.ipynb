```python
# ライブラリの読み込み
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# sklearnからボストンの価格をimport
from sklearn.datasets import load_boston
```

```python
# どのようなものが入っているか
boston = load_boston()
```

```python
# 変数の数が１３個
print(boston.DESCR)
```

### <font color='blue'>まずは可視化</font>

```python
# 教師データをプロットする
plt.hist(boston.target, bins=50)
plt.xlabel('Price ($1,000)')
plt.ylabel('Number of houses')
```

### <font color='blue'>部屋の数と住宅の価格について散布図を描く</font>

```python
# 散布図を描く
# x軸が部屋の数, 縦軸が値段という風になる
plt.scatter(boston.data[:,5], boston.target)
plt.xlabel('Number of Rooms')
plt.ylabel('Price($1,000)')
```

```python
# DataFrameに変換する
boston_df = DataFrame(boston.data)
```

```python
# 列名がついていない
boston_df
```

```python
# 列名をつける
boston_df.columns = boston.feature_names
```

```python
# 教師データをDataFrameに付け加える
boston_df['Price'] = boston.target
```

```python
# 上から10個みる
boston_df.head(10)
```

```python
# 回帰曲線を描く
sns.lmplot('RM', 'Price', data = boston_df)
```

### <font color='blue'>ここまでで、データの概要を把握してヒストグラムを描画した。</font>

numpyを用いて回帰直線の式を求める。<sbr>

```python
X =boston_df.RM
```

```python
X.shape
```

```python
# 配列の形を変える
np.vstack(boston_df.RM).shape
```

```python
# 多次配列に変換
X = np.vstack(boston_df.RM)
```

```python
Y = boston_df.Price
```

```python
[ [value, 1] for value in X ]
```

```python
X = np.array([[value, 1] for value in X])
```

```python
Y.shape
```

```python
X.shape
```

```python
# 回帰直線の係数を求めた
a, b = np.linalg.lstsq(X, Y)[0]
```

```python
# 回帰直線を引いた
plt.plot(boston_df.RM, boston_df.Price, 'o')

x = boston_df.RM
plt.plot(x, a*x+b, 'r')
```

### <font color='blue'>以下からは</font>

```python
result = np.linalg.lstsq(X, Y)
```

```python
# １番目の値が全ての誤差の値となっている
result
```

```python
# したがって
error_total = result[1]
```

```python
rmse = np.sqrt(error_total/len(X))
```

```python
rmse
```

```python
print('平均二乗誤差の平方根={:0.2f}'.format(rmse[0]))
```

### <font color='blue'>重回帰分析に入る</font>

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
lreg = LinearRegression()
```

```python
X_multi = boston_df.drop('Price', 1)
```

```python
X_multi
```

```python
# 13個の説明変数
X_multi.shape
```

```python
Y_target = boston_df.Price
```

```python
lreg.fit(X_multi, Y_target)
```

```python
lreg.intercept_
```

```python
coeff_df = DataFrame(boston_df.columns)
```

```python
coeff_df
```

```python
# 特徴量
coeff_df.columns = ['Features']
```

```python
# 推定された各特徴量を代入
coeff_df['Coefficient Estimate'] = pd.Series(lreg.coef_)
```

```python
# 重回帰で推定された係数
coeff_df
```

```python
import sklearn.cross_validation

# 学習データとテストデータに分ける
X_train, X_test, Y_train, Y_test = sklearn.cross_validation.train_test_split(X_multi, boston_df.Price)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
lreg = LinearRegression()
```

```python
# 学習する
lreg.fit(X_train, Y_train)
```

```python
pred_train = lreg.predict(X_train)
```

```python
pred_test = lreg.predict(X_test)
```

```python
# 平均２乗誤差
np.mean((Y_train - pred_train) ** 2)
```

```python
np.mean((Y_test - pred_test) **2 )
```

```python
# 学習セットの予測された値
train = plt.scatter(pred_train, (pred_train - Y_train), c = 'b', alpha=0.5)
# テストセットの予測された値
test = plt.scatter(pred_test, (pred_test - Y_test), c = 'r', alpha=0.5)
# 誤差が全くないところについてみる
plt.hlines(y=0, xmin=-1.0, xmax=50)

plt.legend((train, test), ('Training', 'Test'), loc = 'lower left')
plt.title('Residual Plots')
```
