まずは変数を１つだけ使って、機械学習を行う。<br>

```python
# ライブラリの読み込み
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# sklearnからボストンの価格をimport
from sklearn.datasets import load_boston
```

```python
# どのようなものが入っているか
boston = load_boston()
```

```python
# 変数の数が１３個
print(boston.DESCR)
```

### <font color='blue'>まずは可視化</font>

```python
# 教師データをプロットする
plt.hist(boston.target, bins=50)
plt.xlabel('Price ($1,000)')
plt.ylabel('Number of houses')
```

### <font color='blue'>部屋の数と住宅の価格について散布図を描く</font>

```python
# 散布図を描く
# x軸が部屋の数, 縦軸が値段という風になる
plt.scatter(boston.data[:,5], boston.target)
plt.xlabel('Number of Rooms')
plt.ylabel('Price($1,000)')
```

```python
# DataFrameに変換する
boston_df = DataFrame(boston.data)
```

```python
# 列名がついていない
boston_df
```

```python
# 列名をつける
boston_df.columns = boston.feature_names
```

```python
# 教師データをDataFrameに付け加える
boston_df['Price'] = boston.target
```

```python
# 上から10個みる
boston_df.head(10)
```

```python
# 回帰曲線を描く
sns.lmplot('RM', 'Price', data = boston_df)
```

### <font color='blue'>ここまでで、データの概要を把握してヒストグラムを描画した。</font>

numpyを用いて回帰直線の式を求める。<sbr>

```python
X =boston_df.RM
```

```python
X.shape
```

```python
# 配列の形を変える
np.vstack(boston_df.RM).shape
```

```python
# 多次配列に変換
X = np.vstack(boston_df.RM)
```

```python
Y = boston_df.Price
```

```python
[ [value, 1] for value in X ]
```

```python
X = np.array([[value, 1] for value in X])
```

```python
Y.shape
```

```python
X.shape
```

```python
# 回帰直線の係数を求めた
a, b = np.linalg.lstsq(X, Y)[0]
```

```python
# 回帰直線を引いた
plt.plot(boston_df.RM, boston_df.Price, 'o')

x = boston_df.RM
plt.plot(x, a*x+b, 'r')
```
